```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

The IMU data zip file is already uploaded, the data path can be changed into the exact filepath of compressed IMU data

```python
data_path = "/Users/daikexin/pythonProject/CompensatoryMovement/data"
```

```python
# decompress the folder of data
import zipfile

# Define the ZIP file path and extraction directory
zip_file_path = f"{data_path}/IMU Data.zip" 
extraction_dir = f"{data_path}"  

# Create the extraction directory if it doesn't exist
os.makedirs(extraction_dir, exist_ok=True)

# Extract the ZIP file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extraction_dir)

print(f"ZIP file extracted to {extraction_dir}")
```

Inorder to  match the participant name in the files

```python
# rename the folder of each participant
base_dir = f"{data_path}/IMU Data"  

for folder_name in os.listdir(base_dir):
    if folder_name in ["P02", "P03", "P04", "P05", "P06"]:
        new_folder_name = folder_name.replace("P02", "P_02").replace("P03", "P_03").replace("P04", "P_04").replace("P05", "P_05").replace("P06", "P_06")
        
        # Rename the folder
        old_path = os.path.join(base_dir, folder_name)
        new_path = os.path.join(base_dir, new_folder_name)
        os.rename(old_path, new_path)
```

import all the meaningful IMU data into one dataframe

```python
# Define the path to the main IMU data folder
imu_path = data_path + "/IMU Data"

condition_task_mapping = {
    '02': ('natural', 'Cupplacing'),
    '03': ('natural', 'Pouring'),
    '04': ('natural', 'Pin'),
    '05': ('natural', 'Wiping'),
    '06': ('comp', 'Cupplacing'),
    '07': ('comp', 'Pouring'),
    '08': ('comp', 'Pin'),
    '09': ('comp', 'Wiping'),
    '10': ('comp_WE', 'Cupplacing'),
    '11': ('comp_WE', 'Pouring'),
    '12': ('comp_WE', 'Pin'),
    '13': ('comp_WE', 'Wiping')
}

body_parts = ['arm_l', 'arm_r', 'trunk', 'wrist_l', 'wrist_r']

def load_imu_data(imu_path, body_parts, condition_task_mapping):
    all_data = []

    for participant_folder in sorted(os.listdir(imu_path)):
        participant_path = os.path.join(imu_path, participant_folder)
        if not os.path.isdir(participant_path):
            continue

        participant_id = participant_folder

        for filename in sorted(os.listdir(participant_path)):
            if filename.endswith('.csv'):
                # Exclude files with '_01_' directly
                if '_01_' in filename:
                    print(f"Excluding file with '_01_': {filename}")
                    continue

                # Split the filename
                parts = filename.split('_')

                # Ensure the filename format is as expected
                if len(parts) < 4:
                    print(f"Skipping file with unexpected format: {filename}")
                    continue

                # Determine the body part (single-word or two-word like "arm_l", "wrist_r")
                if f"{parts[0]}_{parts[1]}" in body_parts:
                    body_part = f"{parts[0]}_{parts[1]}"
                elif parts[0] in body_parts:
                    body_part = parts[0]
                else:
                    print(f"Skipping file due to unmatched body part: {filename}")
                    continue

                # Extract condition_task_code
                condition_task_code = parts[-2]  # Assuming this is the second last part of the filename
                if condition_task_code not in condition_task_mapping:
                    print(f"Skipping file due to unmatched condition-task code: {condition_task_code}")
                    continue

                # Map condition-task code to condition and task
                condition, task = condition_task_mapping[condition_task_code]

                # Debugging: Print the parsed metadata
                print(f"File: {filename} | Participant: {participant_id} | Body Part: {body_part} | Condition: {condition} | Task: {task}")

                # Load the CSV file
                file_path = os.path.join(participant_path, filename)
                data = pd.read_csv(file_path)

                # Add metadata columns
                data.insert(0, 'participant', participant_id)
                data.insert(1, 'body_part', body_part)
                data.insert(2, 'condition', condition)
                data.insert(3, 'task', task)
                data.insert(4, 'file_path', file_path)

                all_data.append(data)

    # Combine all data into a single DataFrame
    if all_data:
        combined_df = pd.concat(all_data, ignore_index=True)
    else:
        combined_df = pd.DataFrame()  # Return an empty DataFrame if no valid files are found

    # Debugging: Check the unique values in the combined DataFrame
    print("Unique tasks per body part and condition in combined DataFrame:")
    print(combined_df.groupby(['body_part', 'condition', 'task']).size())

    return combined_df

# Reload the IMU data
imu_df = load_imu_data(imu_path, body_parts, condition_task_mapping)
```
